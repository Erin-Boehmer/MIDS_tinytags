```python
################################################
### --- Assumes [rrr...ggg...bbb] format --- ###
################################################

### --- Necessary Setup --- ###
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

### --- Read in the JSON --- ###
with open("imageSample.json") as json_file:
    json_data = json.load(json_file)
    
### --- Pick out the relevant field and convert from Unicode to Numpy Array--- ###
image = str(json_data['data']) #extract relevant info from the json and parse as string
image = image[1:-1].split(",") #drop off the front and back brackets from string list and split into an array
image_clean = np.asarray([(255-int(x)) for x in image]) #reverse the scale and parse as nparray
proper_image = image_clean.reshape(3,1024).swapaxes(0,1).reshape(32,32,3) #reshape data to plot with plt

### --- Display--- ###
plt.imshow(proper_image)
plt.show()
```
